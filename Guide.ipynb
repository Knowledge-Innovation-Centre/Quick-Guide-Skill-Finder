```python
import requests
from pprint import pprint
```

```python
url = "https://find.skilldata.info/send_query_full"
```

```python
headers = {
    "Content-Type": "application/json",
    "KIC-API-HEADER": "YOUR_API_KEY"  # Replace with your actual API key
}
```

```python
# Defaults to return 12 search results
data = {
    "skill_input": "Data Science",
    "language": "en"
}
```

```python
# Manually set new return value for the search results
data = {
  "skill_input": "Data Science",
  "limit": 16,
  "language": "de", #de for German
}
```

```python
response = requests.post(url, json=data, headers=headers)
```

```python
pprint(response.json())
```
